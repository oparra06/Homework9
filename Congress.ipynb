```python
for congressperson in doc.find_all("li", class_="compact"):
    result = congressperson.find("a")
    name = result.text
    url = "https://www.congress.gov/members" + (result.get('href'))
    state = congressperson.find(text="State:").next.next.next
    party = congressperson.find(text="Party:").next.next.next
    served = congressperson.find(text="Served:").next.next.text
    print(name, url, state, party, served)

congress_all = []
for congressperson in doc.find_all("li", class_="compact"):
    result = congressperson.find("a")
    name = result.text
    url = "https://www.congress.gov/members" + (result.get('href'))
    state = congressperson.find(text="State:").next.next.next
    party = congressperson.find(text="Party:").next.next.next
    served = congressperson.find(text="Served:").next.next.text
    congressperson = {
        'name' : name,
        'url' : url,
        'state' : state,
        'party' : party,
        'served' : served
    }
    congress_all.append(congressperson)
df = pd.DataFrame(congress_all)
df.to_csv("congress.csv")
df
```
