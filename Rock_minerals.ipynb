```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
response = requests.get("https://rocktumbler.com/blog/rock-and-mineral-clubs/")
doc = BeautifulSoup(response.text)
doc
```

```python
stories = []
impar = -1;
for tb in all_table[1:51]:    
    impar = impar + 2
    table = doc.find_all("table")[impar]
    for row in table.find_all('tr'):
        name = (row.find("a").text)
        url = (row.find("a")["href"])
        city = (row.find_all("td")[1].text)
        story = {
        'name' : name,
        'url': url,
        'city' : city
    }
    
    stories.append(story)

df = pd.DataFrame(stories)
df
    
```

```python
df.to_csv("rocks_minerals.csv")
```
